```python
from pathlib import Path
import openpyxl
from openpyxl.utils.cell import get_column_letter
import os
import tkinter as tk
from tkinter import ttk
from tkinter.filedialog import askdirectory
from PIL import ImageTk, Image
```

```python
class ExcelMonkey:
    def disableSheet(event):
            #nastavenie stavu edit/disable pre entry policko Excel sheetname
            
            if (event.widget.get())== 'Change Excel filename':
                Sheetfld['state'] = 'disabled'
            else:
                Sheetfld['state'] = 'normal'
            
    def selectDirectory():
        #vyber foldra kde chceme spravit zmeny
        filepath = tk.filedialog.askdirectory(title='Select your folder') #Shows dialog box and return the path
        folderPath.set(filepath)
        

    def changeFilename():
                # zmena nazvu vsetkych Excel suborov vo foldri
                
                                
                Find=str(Findfld.get())
                Replace=str(Replacefld.get())
                logFrame=tk.Frame(window)
                logFrame.place(x=10,y=400)
                scrollbar = tk.Scrollbar(logFrame)
                log_widget = tk.Text(logFrame, height=25, width=120, yscrollcommand=scrollbar.set)
                scrollbar.config(command=log_widget.yview)
                scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
                log_widget.pack(side="left")
                total=0
                path_to_xls = folderPath.get()
                
                
                for xls in os.listdir(path_to_xls): 
                    print(xls)     
                    try:
                        if xls.endswith(".xls") or xls.endswith(".xlsx"):
                            
                            newName= xls.replace(Find, Replace)
                            os.rename(os.path.join(path_to_xls,xls),os.path.join(path_to_xls,newName)) 
                            log_widget.insert(tk.END, ("OK      " + newName+ "\n")) 
                               
                    except Exception as e:  
                            total+=1
                            error=str(e)
                            log_widget.insert(tk.END, ("ERROR   " + xls + " ---- " +error + "\n")) #vytvorenie logu pre neuspesnu zmenu
                
                #hlaska po dokonceni behu programu 
                done=tk.Label(window, text="Done !", fg='green', font=("Comic sans MS", 14))
                done.place(x=220, y=80)
                total_error=tk.Label(window, text="Total errors = " + str(total), fg='red', font=("Comic sans MS", 10))
                total_error.place(x=190, y=120)        
    
      
          
        
        
    #funkcia po odkliknuti tlacitka submit
    
    def click():

                Sheet=str(Sheetfld.get())
                Find=str(Findfld.get())
                Replace=str(Replacefld.get())
                logFrame=tk.Frame(window)
                logFrame.place(x=10,y=400)
                scrollbar = tk.Scrollbar(logFrame)
                log_widget = tk.Text(logFrame, height=25, width=120, yscrollcommand=scrollbar.set)
                scrollbar.config(command=log_widget.yview)
                scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
                log_widget.pack(side="left")
                total=0
                path_to_xls = folderPath.get()
                change = selectboxChange.get()
                
                
                if  change == "Change Excel sheet":

               
                    for xls in os.listdir(path_to_xls): 
                        
                        try:
                            if xls.endswith(".xls") or xls.endswith(".xlsx"):
                                    
                                wb = openpyxl.load_workbook((os.path.join(path_to_xls,xls)))
                                wb.sheetnames
                                sheet = wb[Sheet]
                                number_rows = sheet.max_row
                                number_columns = sheet.max_column

                                original_merged_cells=sheet.merged_cell_ranges
                                number_merged=(len(original_merged_cells))
                                

                               #zrusime vsetky merged cells v exceli
                                i=0
                                while i < number_merged: 
                                    
                                    sheet.unmerge_cells(str(sheet.merged_cell_ranges[i]))
                                    merged_cells=sheet.merged_cell_ranges
                                    number_merged=(len(merged_cells))
                                    


                                #Zmena textu v bunkach excelu
                                for r in range(number_columns):
                                    for k in range(number_rows):

                                        cell = str(sheet[get_column_letter(r+1)+str(k+1)].value)
                                      
                                        oprava=cell.replace(Find, Replace)
                                        type(oprava)
                                        if oprava == "None":
                                                    oprava =""

                                        sheet[get_column_letter(r+1)+str(k+1)] = oprava
                                        

                                #opatovne vytvorenie merged cells ako boli v orinalnom exceli    
                                number_merged=(len(original_merged_cells))
                                i=0
                                while i < number_merged: 
                                    

                                    sheet.merge_cells(str(original_merged_cells[i]))
                                    i+=1
                                    
                                log_widget.insert(tk.END, ("OK      " + xls+ "\n"))  #vytvorenie logu pre uspesnu zmenu
                                wb.save((os.path.join(path_to_xls,xls)))
                                wb.close()
                        except Exception as e:
                            total+=1
                            error=str(e)
                            log_widget.insert(tk.END, ("ERROR   " + xls + " ---- " +error + "\n")) #vytvorenie logu pre neuspesnu zmenu


                    #hlaska po dokonceni behu programu            
                    done=tk.Label(window, text="Done !", fg='green', font=("Comic sans MS", 14))
                    done.place(x=220, y=80)
                    total_error=tk.Label(window, text="Total errors = " + str(total), fg='red', font=("Comic sans MS", 10))
                    total_error.place(x=190, y=120)
                else:
                    ExcelMonkey.changeFilename()
                 
    
window = tk.Tk()   
window.title('Excel Monkey v2')
window.geometry("1000x825")

canvas = tk.Canvas(window, width =260, height = 200) 

canvas.place(x=550, y=50)  
im = Image.open('opicka.png')
img = im.resize((380, 260))
ph = ImageTk.PhotoImage(img)

#c = tk.Canvas(canvas)
#c.pack()
#c.place(x=150, y=30)
canvas.create_image(80, 100, image=ph)

folderPath = tk.StringVar()

name=tk.Label(window, text="Excel Monkey", fg='brown', font=("Magneto", 22))
name.place(x=40, y=20) 
copyright=tk.Label(window, text="\u00a9 aviktory", fg='black', font='none 10')
copyright.place(rely=1.0, relx=1.0, x=0, y=0, anchor=tk.SE)   
lbl=tk.Label(window, text="Software for text replacement in multiple Excel files for TQM monkeys", fg='black', font=("Comic sans MS", 9))
lbl.place(x=40, y=60)
#help_txt=tk.Label(window, text="Help: Copy Excel Monkey to folder with your Excel files", fg='black', font='none 10')
#help_txt.place(rely=1.0, relx=0.0, x=0, y=0, anchor=tk.SW)
submit=tk.Button(window, text="Submit", fg='green',font='sans 8 bold', command=ExcelMonkey.click, height =2, width=15)
submit.place(x=50, y=350)
directory=tk.Button(window, text="Select directory", fg='brown',font='sans 8 bold', command=ExcelMonkey.selectDirectory, height =1, width=15)
directory.place(x=50, y=310)
entry=tk.Entry(window,textvariable=folderPath, width=100)
entry.place(x=180, y=310)
Sheetfld=tk.Entry(window, fg='blue',width=23)
Sheetfld.place(x=180, y=190)
Findfld=tk.Entry(window, fg='blue',width=23)
Findfld.place(x=180, y=230)
Replacefld=tk.Entry(window, fg='blue',width=23)
Replacefld.place(x=180, y=270)

Sheetlbl=tk.Label(window, text="Excel sheet name")
Findlbl=tk.Label(window, text="Find what")
Replacelbl=tk.Label(window, text="Replace with")
Sheetlbl.place(x=50, y=190)
Findlbl.place(x=50, y=230)
Replacelbl.place(x=50, y=270)

selectChange = tk.StringVar()
selectboxChange=ttk.Combobox(window, textvariable = selectChange)
selectboxChange.place(x=180, y=150)
selectChangelabel = ttk.Label(text="Select change")
selectChangelabel.place(x=50, y=150)

selectboxChange['values']=('Change Excel sheet', 'Change Excel filename')
selectboxChange['state'] = 'readonly'
selectChange.set('Change Excel sheet')
selectboxChange.bind("<<ComboboxSelected>>", ExcelMonkey.disableSheet)


window.mainloop()
```
